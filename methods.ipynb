# Methods

## 1. Nonequilibrium Green's Function (NEGF) Formalism

To investigate the nonlinear transport properties of MnBi$_2$Te$_4$ in the presence of strong disorder and Anderson localization, we employ the Nonequilibrium Green's Function (NEGF) formalism. While the Kubo formula is widely used for linear response in clean or weakly disordered systems, the NEGF approach offers several critical advantages for our study:

1.  **Large-Scale Disordered Systems:** We simulate large supercells (up to $400 \times 400$ sites) to capture the localization length $\xi$ near the critical point. Full diagonalization required for the Kubo formula is computationally prohibitive ($\mathcal{O}(N^3)$) for such system sizes. The recursive Green's function (RGF) algorithm allows us to compute the necessary transport quantities with $\mathcal{O}(N)$ scaling in the transport direction.
2.  **Open Boundary Conditions:** NEGF naturally incorporates open boundary conditions via self-energies, which is essential for simulating realistic transport setups and avoiding finite-size artifacts associated with periodic boundary conditions in the presence of strong disorder.
3.  **Anderson Localization:** The formalism directly computes the transmission probability, which is the defining quantity for the metal-insulator transition and Anderson localization, without relying on assumptions about the nature of the eigenstates.

The central quantity is the retarded Green's function of the scattering region:
$$ G^R(E) = [E \mathbb{I} - H_{C} - \Sigma_L(E) - \Sigma_R(E)]^{-1} $$
where $H_C$ is the Hamiltonian of the central disordered region, and $\Sigma_{L,R}(E)$ are the self-energies of the semi-infinite left and right leads, calculated using the Sancho-Rubio iterative method.

## 2. Current Calculation and Finite Difference Method

### 2.1 Local Current Distribution
Instead of relying on diagrammatic expansions for the second-order response, which can be cumbersome and prone to error in disordered systems, we adopt a **Finite Difference Method** based on the direct calculation of current under finite bias.

We apply a longitudinal bias voltage $V_x$ across the sample, setting the chemical potentials of the leads as $\mu_L = E_F + eV_x/2$ and $\mu_R = E_F - eV_x/2$. The local bond current flowing from site $\mathbf{i}$ to $\mathbf{j}$ is calculated using the Keldysh formalism:
$$ J_{\mathbf{i} \to \mathbf{j}} = \frac{2e}{\hbar} \int \frac{dE}{2\pi} \text{Re} \left[ \text{Tr} \left( H_{\mathbf{i}\mathbf{j}} G^<_{\mathbf{j}\mathbf{i}}(E) \right) \right] $$
where $G^<(E) = G^R(E) \Sigma^<(E) G^A(E)$ is the lesser Green's function, and $\Sigma^<(E) = i \sum_{\alpha=L,R} f_\alpha(E) \Gamma_\alpha(E)$. The total transverse current $I_y$ is obtained by summing the appropriate bond currents across a cross-section of the sample.

### 2.2 Extraction of Second-Order Nonlinear Conductance
The transverse current response $I_y$ can be expanded in powers of the applied longitudinal voltage $V_x$:
$$ I_y(V_x) = G_{xy}^{(1)} V_x + G_{xy}^{(2)} V_x^2 + \mathcal{O}(V_x^3) $$
where $G_{xy}^{(1)}$ is the linear Hall conductance and $G_{xy}^{(2)}$ is the second-order nonlinear Hall conductance.

To extract the nonlinear coefficient $G_{xy}^{(2)}$, we perform two independent transport calculations with reversed bias polarities, $+V_x$ and $-V_x$. By summing the results, the linear term (odd in $V_x$) cancels out, isolating the second-order contribution (even in $V_x$):
$$ I_y(+V_x) + I_y(-V_x) \approx 2 G_{xy}^{(2)} V_x^2 $$

Thus, the second-order nonlinear Hall conductance is given by:
$$ G_{xy}^{(2)} \approx \frac{I_y(+V_x) + I_y(-V_x) - 2I_y(0)}{2 V_x^2} $$
Here, $I_y(0)$ represents the equilibrium current, which vanishes theoretically but is included to minimize numerical noise. We choose a small bias $V_x$ within the linear regime of the longitudinal current but large enough to resolve the nonlinear transverse signal above numerical precision.

## 3. Disentanglement of BCD and QMD Contributions

The nonlinear Hall effect in topological materials can originate from two distinct geometric quantities: the Berry Curvature Dipole (BCD) and the Quantum Metric Dipole (QMD). To distinguish these mechanisms near the Anderson critical point, we utilize their distinct symmetry properties under the inversion of an out-of-plane electric field $E_z$ (simulated by an on-site potential difference or sublattice symmetry breaking term).

*   **Berry Curvature Dipole (BCD):** The BCD is a pseudovector and is odd under inversion symmetry breaking. Its contribution to the nonlinear conductivity, $\sigma_{BCD}^{(2)}$, reverses sign when the symmetry-breaking field $E_z$ is reversed.
*   **Quantum Metric Dipole (QMD):** The QMD contribution, $\sigma_{QMD}^{(2)}$, is typically even under the reversal of $E_z$ in this setup.

We can therefore separate the two components by calculating the nonlinear conductance at $+E_z$ and $-E_z$:

$$ \sigma_{BCD}^{(2)} = \frac{\sigma_{xy}^{(2)}(+E_z) - \sigma_{xy}^{(2)}(-E_z)}{2} $$
$$ \sigma_{QMD}^{(2)} = \frac{\sigma_{xy}^{(2)}(+E_z) + \sigma_{xy}^{(2)}(-E_z)}{2} $$

This separation allows us to unambiguously identify the QMD as the driver of the "nonlinear Hall catastrophe" at the topological phase transition, where the quantum metric diverges due to the fractal nature of the critical wavefunctions.
