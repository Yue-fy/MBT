## 1. 层霍尔效应简介 / Introduction to Layer Hall Effect

### 物理图像 / Physical Picture

**层霍尔效应（Layer Hall Effect, LHE）** 是反铁磁拓扑绝缘体中的一种新型输运现象：

- **输入**：平面内电场 $\mathbf{E}_{\parallel}$（例如沿 $y$ 方向）
- **输出**：横向的**层分辨电流** $j_x^{(l)}$（不同层相反）

$$
j_x^{(l)} = \sigma_{xy}^{(l)} E_y
$$

**关键特点**：
- 总电流 $j_x^{\text{total}} = \sum_l j_x^{(l)} = 0$（反铁磁对称性）
- 层反对称电流 $j_x^{\text{layer}} = j_x^{(1)} - j_x^{(2)} \neq 0$

### 物理起源 / Physical Origin

层霍尔效应源于：
1. **层分辨 Berry 曲率** $\Omega_n^{(l)}(\mathbf{k})$
2. **反铁磁序**：上下层磁矩反平行
3. **层间耦合**：决定层间相干性

### MnBi₂Te₄ 材料

- **晶体结构**：七层 Te-Bi-Te-Mn-Te-Bi-Te 重复单元
- **磁结构**：层间反铁磁耦合（A型）
- **拓扑性质**：轴子绝缘体（$\theta = \pi$）
- **实验观测**：Gao et al., *Nature* **595**, 521 (2021)

## 2. Berry 曲率理论 / Berry Curvature Theory

### 2.1 Bloch 态和 Berry 联络

对于周期性晶体，Bloch 态为：
$$
|\psi_{n\mathbf{k}}\rangle = e^{i\mathbf{k}\cdot\mathbf{r}}|u_{n\mathbf{k}}\rangle
$$

**Berry 联络**（momentum space gauge field）：
$$
\mathcal{A}_{n}^{\mu}(\mathbf{k}) = i\langle u_{n\mathbf{k}}|\frac{\partial}{\partial k_{\mu}}|u_{n\mathbf{k}}\rangle
$$

### 2.2 Berry 曲率（核心公式）

Berry 曲率是动量空间的
：
$$
\Omega_{n}^{\mu\nu}(\mathbf{k}) = \partial_{k_\mu}\mathcal{A}_{n}^{\nu} - \partial_{k_\nu}\mathcal{A}_{n}^{\mu}
$$

对于二维系统：
$$
\Omega_n(\mathbf{k}) = \Omega_n^{xy}(\mathbf{k})
$$

### 2.3 Kubo 公式（数值计算用）

**规范不变的表达式**：
$$
\boxed{
\Omega_n(\mathbf{k}) = -2\text{Im}\sum_{m\neq n}\frac{\langle u_{n\mathbf{k}}|v_x|u_{m\mathbf{k}}\rangle\langle u_{m\mathbf{k}}|v_y|u_{n\mathbf{k}}\rangle}{(E_n(\mathbf{k}) - E_m(\mathbf{k}))^2}
}
$$

其中速度算符：
$$
v_\mu = \frac{1}{\hbar}\frac{\partial \mathcal{H}}{\partial k_\mu}
$$

### 2.4 层分辨 Berry 曲率

对于多层系统，定义层投影算符 $\hat{P}_l$，层分辨 Berry 曲率为：
$$
\boxed{
\Omega_n^{(l)}(\mathbf{k}) = -2\text{Im}\sum_{m\neq n}\frac{\langle u_{n\mathbf{k}}|\hat{P}_l v_x|u_{m\mathbf{k}}\rangle\langle u_{m\mathbf{k}}|v_y|u_{n\mathbf{k}}\rangle}{(E_n - E_m)^2}
}
$$

### 2.5 Chern 数

拓扑不变量：
$$
C_n = \frac{1}{2\pi}\int_{\text{BZ}} d^2\mathbf{k}\, \Omega_n(\mathbf{k})
$$

量子反常霍尔电导率：
$$
\sigma_{xy} = C \frac{e^2}{h}
$$

## 3. 量子度规 / Quantum Metric

### 3.1 定义

**量子度规张量**是量子几何张量的实部：
$$
\boxed{
g_{n}^{\mu\nu}(\mathbf{k}) = \text{Re}\sum_{m\neq n}\frac{v_{nm}^{\mu}(v_{nm}^{\nu})^*}{(E_n - E_m)^2}
}
$$

其中 $v_{nm}^{\mu} = \langle u_{n\mathbf{k}}|v_\mu|u_{m\mathbf{k}}\rangle$

### 3.2 物理意义

- **几何距离**：测量 $\mathbf{k}$ 空间中 Bloch 态的

- **能带平坦度**：平带 → 大的量子度规
- **波函数局域化**：与实空间定域性相关

### 3.3 与 Berry 曲率的关系

量子几何张量：
$$
Q_{n}^{\mu\nu} = g_{n}^{\mu\nu} + \frac{i}{2}\Omega_{n}^{\mu\nu}
$$

- **实部**：量子度规 $g$ → 影响超流权重、非线性光学
- **虚部**：Berry 曲率 $\Omega$ → 霍尔效应

### 3.4 量子度规对层霍尔的贡献

在掺杂或半金属系统中，量子度规偶极子贡献：
$$
\sigma_{xy}^{\text{metric}} \sim e^2 \tau \sum_n \int d^2k\, \left(-\frac{\partial f}{\partial \epsilon}\right) \frac{\partial E_n}{\partial k_x} \frac{\partial g_n^{\text{layer}}}{\partial k_y}
$$

其中 $\tau$ 是散射时间。

## 4. MnBi₂Te₄ 紧束缚模型

### 4.1 双层哈密顿量

考虑有效双层模型（简化的 MnBi₂Te₄）：
$$
\mathcal{H}(\mathbf{k}) = \begin{pmatrix}
\mathcal{H}_{11}(\mathbf{k}) & \mathcal{H}_{12}(\mathbf{k}) \\
\mathcal{H}_{21}(\mathbf{k}) & \mathcal{H}_{22}(\mathbf{k})
\end{pmatrix}
$$

### 4.2 层内哈密顿量

$$
\mathcal{H}_{ll}(\mathbf{k}) = h_0(\mathbf{k}) \otimes \mathbb{I}_{\sigma} + \mathbf{h}_{\text{SOC}}(\mathbf{k}) \cdot \boldsymbol{\sigma} + M_l \sigma_z
$$

**动能项**（三角格子最近邻）：
$$
h_0(\mathbf{k}) = -2t \sum_{i=1}^{3} \cos(\mathbf{k}\cdot\boldsymbol{\delta}_i)
$$

最近邻矢量：
$$
\boldsymbol{\delta}_1 = a(1,0), \quad \boldsymbol{\delta}_2 = a\left(-\frac{1}{2}, \frac{\sqrt{3}}{2}\right), \quad \boldsymbol{\delta}_3 = a\left(-\frac{1}{2}, -\frac{\sqrt{3}}{2}\right)
$$

**自旋轨道耦合**（Kane-Mele 型）：
$$
\mathbf{h}_{\text{SOC}}(\mathbf{k}) = \lambda_{\text{SO}} \sum_{i=1}^{3} \sin(\mathbf{k}\cdot\boldsymbol{\delta}_i) (\hat{z} \times \boldsymbol{\delta}_i)
$$

**交换场**：
$$
M_1 = +M, \quad M_2 = -M \quad \text{(反铁磁)}
$$

### 4.3 层间耦合

$$
\mathcal{H}_{12}(\mathbf{k}) = t_{\perp}(\mathbf{k}) \mathbb{I}_{\sigma}
$$

$$
t_{\perp}(\mathbf{k}) = t_{\perp}^{(0)} + t_{\perp}^{(1)} \sum_{i=1}^3 \cos(\mathbf{k}\cdot\boldsymbol{\delta}_i)
$$

### 4.4 完整基矢

$$
|\psi\rangle = \begin{pmatrix}
c_{1\uparrow}^\dagger \\ c_{1\downarrow}^\dagger \\ c_{2\uparrow}^\dagger \\ c_{2\downarrow}^\dagger
\end{pmatrix}
\quad \text{(层1上，层1下，层2上，层2下)}
$$

哈密顿量维度：$4\times 4$（2层 × 2自旋）

## 5. 层霍尔电导率计算

### 5.1 本征贡献（Berry 曲率）

零温下：
$$
\boxed{
\sigma_{xy}^{\text{layer}} = -\frac{e^2}{\hbar} \sum_{n\in\text{occ}} \int_{\text{BZ}} \frac{d^2\mathbf{k}}{(2\pi)^2} \Delta\Omega_n(\mathbf{k})
}
$$

其中**层 Berry 曲率对比**：
$$
\Delta\Omega_n(\mathbf{k}) = \Omega_n^{(1)}(\mathbf{k}) - \Omega_n^{(2)}(\mathbf{k})
$$

### 5.2 有限温度

$$
\sigma_{xy}^{\text{layer}} = -\frac{e^2}{\hbar} \sum_{n} \int_{\text{BZ}} \frac{d^2\mathbf{k}}{(2\pi)^2} f(E_n(\mathbf{k}), \mu, T) \Delta\Omega_n(\mathbf{k})
$$

Fermi-Dirac 分布：
$$
f(E, \mu, T) = \frac{1}{1 + e^{(E-\mu)/(k_B T)}}
$$

### 5.3 数值积分

离散 k 网格：
$$
\sigma_{xy}^{\text{layer}} \approx -\frac{e^2}{\hbar(2\pi)^2} \sum_{n\in\text{occ}} \sum_{\mathbf{k}_i} \Delta\Omega_n(\mathbf{k}_i) \Delta k_x \Delta k_y
$$

其中 $\Delta k_x = \Delta k_y = \frac{2\pi}{a \cdot N_k}$
