# Theoretical Framework: Nonlinear Hall Catastrophe at Anderson Criticality

## 1. Introduction
This notebook details the theoretical construction of the effective lattice model for the magnetic topological insulator MnBi$_2$Te$_4$ and the Non-Equilibrium Green's Function (NEGF) formalism used to calculate the nonlinear transport properties. We start from the continuous $k \cdot p$ Hamiltonian, perform a rigorous lattice regularization, and then introduce the transport methodology.

## 2. Continuous $k \cdot p$ Model
The low-energy physics of MnBi$_2$Te$_4$ thin films is described by a 4-band effective Hamiltonian near the $\Gamma$ point. Based on the $D_{3d}$ point group symmetry, the generic form is:

$$ H(\mathbf{k}) = M(\mathbf{k}) \tau_z \sigma_0 + v_F (k_x \tau_x \sigma_x + k_y \tau_y \sigma_y) $$

where:
*   $\mathbf{k} = (k_x, k_y)$ is the momentum vector.
*   $\tau$ and $\sigma$ are Pauli matrices acting on the orbital and spin subspaces, respectively.
*   $M(\mathbf{k}) = m - B(k_x^2 + k_y^2)$ is the mass term with band inversion parameters $m$ and $B$.
*   $v_F$ is the Fermi velocity.

## 3. Lattice Regularization (Discretization)
To perform numerical simulations on a finite grid, we regularize the continuous model onto a square lattice with lattice constant $a$. We use the standard finite-difference substitutions:

1.  **Linear terms ($k_i$):**
    $$ k_i \to \frac{1}{a} \sin(k_i a) $$
2.  **Quadratic terms ($k_i^2$):**
    $$ k_i^2 \to \frac{2}{a^2} [1 - \cos(k_i a)] $$

Substituting these into $H(\mathbf{k})$ yields the lattice Hamiltonian:

$$ H_{latt}(\mathbf{k}) = \left( m - \frac{2B}{a^2} [2 - \cos(k_x a) - \cos(k_y a)] \right) \tau_z \sigma_0 + \frac{v_F}{a} [\sin(k_x a) \tau_x \sigma_x + \sin(k_y a) \tau_y \sigma_y] $$

## 4. Real-Space Tight-Binding Hamiltonian
Performing an inverse Fourier transform gives the real-space Hamiltonian:
$$ H = \sum_{\mathbf{r}} \Psi_{\mathbf{r}}^\dagger \epsilon_{\mathbf{r}} \Psi_{\mathbf{r}} + \sum_{\mathbf{r}} \left( \Psi_{\mathbf{r}}^\dagger T_x \Psi_{\mathbf{r}+\hat{x}} + \Psi_{\mathbf{r}}^\dagger T_y \Psi_{\mathbf{r}+\hat{y}} + h.c. \right) $$

### 4.1 Hopping Matrices ($T_x, T_y$)
$$ T_x = \frac{B}{a^2} \tau_z \sigma_0 - i \frac{v_F}{2a} \tau_x \sigma_x $$
$$ T_y = \frac{B}{a^2} \tau_z \sigma_0 - i \frac{v_F}{2a} \tau_y \sigma_y $$

### 4.2 On-site Energy ($\epsilon_{\mathbf{r}}$)
$$ \epsilon_{\mathbf{r}} = \left( m - \frac{4B}{a^2} \right) \tau_z \sigma_0 + V_{dis}(\mathbf{r}) $$

### 4.3 Parameter Mapping
Using the fitted parameters for MnBi$_2$Te$_4$ ($v_2=1.41, v_4=1.09, m_0=1.025, m_1=0.975, t_0=0.5, t_1=1.0$):
*   $T_x = \frac{t_0+t_1}{2} \tau_z \sigma_0 - i \frac{v_2+v_4}{2} \tau_x \sigma_x$
*   $\epsilon_{\mathbf{r}} = (m_0 + m_1 - 4t_0 - 4t_1) \tau_z \sigma_0 + V_{dis}(\mathbf{r})$

## 5. Non-Equilibrium Green's Function (NEGF) Formalism

To calculate the transport properties in the presence of disorder and finite bias, we employ the NEGF method. This approach is superior to the Kubo formula for large disordered systems as it naturally handles open boundary conditions and avoids expensive full diagonalization.

### 5.1 Retarded Green's Function
The central quantity is the Retarded Green's function of the scattering region:

$$ G^R(E) = [E \mathbb{I} - H_{C} - \Sigma_L(E) - \Sigma_R(E)]^{-1} $$

where:
*   $H_C$ is the Hamiltonian of the central scattering region.
*   $\Sigma_{L,R}(E)$ are the self-energies describing the coupling to the semi-infinite Left (L) and Right (R) leads. These are calculated iteratively or via the Sancho-Rubio method.

### 5.2 Transmission and Current
The transmission coefficient $T(E)$ from the left to the right lead is given by the Fisher-Lee formula:

$$ T(E) = \text{Tr}[\Gamma_L(E) G^R(E) \Gamma_R(E) G^A(E)] $$

where:
*   $G^A = (G^R)^\dagger$ is the Advanced Green's function.
*   $\Gamma_{\alpha} = i [\Sigma_{\alpha} - \Sigma_{\alpha}^\dagger]$ is the broadening matrix for lead $\alpha \in \{L, R\}$.

The current $I$ flowing through the device under a bias voltage $V$ is:

$$ I = \frac{e}{h} \int_{-\infty}^{\infty} dE \, T(E) [f_L(E) - f_R(E)] $$

where $f_{\alpha}(E) = [1 + \exp((E - \mu_{\alpha})/k_B T)]^{-1}$ is the Fermi-Dirac distribution of lead $\alpha$. For a bias $V_x$ applied along the transport direction, $\mu_L = E_F + eV_x/2$ and $\mu_R = E_F - eV_x/2$.

### 5.3 Calculation of Nonlinear Hall Effect
We are interested in the second-order nonlinear Hall conductance $\sigma_{xy}^{(2)}$. Directly computing the second-order response function using diagrammatic NEGF is complex. Instead, we use a **Finite Difference Method** based on the current response to a finite bias.

**Setup:**
1.  Apply a longitudinal bias voltage $V_x$ between the left and right leads.
2.  Calculate the transverse current $I_y$. In a lattice model, this is done by defining a transverse cross-section and summing the bond currents, or more formally by computing the transmission between top and bottom virtual leads (if using a 4-terminal setup) or analyzing the local current distribution.

**Local Current Operator:**
The bond current between site $\mathbf{i}$ and $\mathbf{j}$ is:
$$ J_{\mathbf{i} \to \mathbf{j}} = \frac{2e}{\hbar} \text{Im} \left[ \text{Tr} \left( H_{\mathbf{i}\mathbf{j}} G^<_{\mathbf{j}\mathbf{i}} \right) \right] $$
where $G^< = G^R \Sigma^< G^A$ is the lesser Green's function, and $\Sigma^< = i \sum_\alpha f_\alpha \Gamma_alpha$.

**Second-Order Conductance:**
The transverse current $I_y$ can be expanded in powers of the longitudinal voltage $V_x$:
$$ I_y(V_x) = G_{xy}^{(1)} V_x + G_{xy}^{(2)} V_x^2 + \mathcal{O}(V_x^3) $$

To extract the second-order coefficient $G_{xy}^{(2)}$, we perform two measurements with reversed bias polarity ($+V_x$ and $-V_x$):

$$ I_y(+V_x) \approx G_{xy}^{(1)} V_x + G_{xy}^{(2)} V_x^2 $$
$$ I_y(-V_x) \approx -G_{xy}^{(1)} V_x + G_{xy}^{(2)} V_x^2 $$

Summing these eliminates the linear term:
$$ I_y(+V_x) + I_y(-V_x) \approx 2 G_{xy}^{(2)} V_x^2 $$

Thus, the nonlinear Hall conductance is:
$$ G_{xy}^{(2)} \approx \frac{I_y(+V_x) + I_y(-V_x) - 2I_y(0)}{2 V_x^2} $$

Note: $I_y(0)$ should be zero in equilibrium, but we include it for numerical stability. This quantity $G_{xy}^{(2)}$ is directly related to the nonlinear Hall conductivity $\sigma_{xy}^{(2)}$ by geometric factors.

## 6. Disorder Implementation
To study the Anderson criticality, we introduce disorder into the Hamiltonian $H_C$:

1.  **Scalar Disorder:** $V_{dis}^{scalar}(\mathbf{r}) \in [-W/2, W/2]$ added to the diagonal elements. This preserves time-reversal symmetry (in the absence of magnetic field) but breaks translational symmetry.
2.  **Magnetic Disorder:** $\delta m_{\mathbf{r}} \tau_z \sigma_0$ added to the mass term. This breaks both time-reversal and translational symmetries locally.

The "Catastrophe" is expected to manifest as a divergence in $G_{xy}^{(2)}$ as the disorder strength $W$ drives the system through a topological phase transition.
